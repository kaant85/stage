```python
from PIL import Image
import numpy as np
import csv

def array_cut(arr):
    tmp_set_gor = set()
    tmp_set_ver = set()
    for i in range(len(arr)):
        for j in range(len(arr[i])):
            if arr[i, j] != 255:
                tmp_set_gor.add(i)
                tmp_set_ver.add(j)   
    return arr[min(tmp_set_gor)-1:max(tmp_set_gor)+2, min(tmp_set_ver)-1:max(tmp_set_ver)+2]


def image_clean(file_name):
    img = Image.open(file_name)
    arr_ = np.asarray(img, dtype='uint8')
    x, y, _ = arr_.shape
    k = np.array([[[0.2989, 0.587, 0.114]]])
    arr2 = np.round(np.sum(arr_*k, axis=2)).astype(np.uint8).reshape((x, y))
    for i in range(len(arr2)):
        for j in range(len(arr2[i])):
            if arr2[i, j] > 150:
                arr2[i, j] = 255
            # elif arr2[i, j] < 70:
            #     arr2[i, j] = 1
            # else:
            #     arr2[i, j] = 1
    img2 = Image.fromarray(array_cut(arr2))
    img2.save('result.png')


def copy_bukva_recurs(ii, jj):
    result = 1
    arr_copy[ii, jj] = arr[ii, jj]
    arr[ii, jj] = 255
    for i in range(ii-1, ii+2):
        for j in range(jj-1, jj+2):
            if arr[i, j] != 255:
                result += copy_bukva_recurs(i, j)
    return result

def get_bukva():
    global arr
    global arr_copy
    global my_list
    my_list = []
    result = 0
    count = 0
    for j in range(len(arr[0])):
        for i in range(len(arr)):
            if arr[i, j] != 255:
                result = copy_bukva_recurs(i, j)
                count += 1
                my_list.append(arr_to_list(array_cut(arr_copy)))
                arr_copy = np.full((len(arr), len(arr[0])), 255, dtype=int)
                
    print(count)
    return result


def arr_to_list(ar):
    res_list = []
    ar_ini = np.full((14, 14), 255, dtype=int)
    for i in range(len(ar)):
        for j in range(len(ar[i])):
            ar_ini[i, j] = ar[i, j]
    for elem in ar_ini:
        res_list.extend(elem.tolist())
    return res_list
    
def save_file(my_list):
    with open('out.csv', 'w') as file:
        writer = csv.writer(file)
        writer.writerows(my_list)



image_clean(input("Введите имя файла: "))

img = Image.open('result.png')
arr = np.asarray(img, dtype=int)
arr_copy = np.full((len(arr), len(arr[0])), 255, dtype=int)
get_bukva()
save_file(my_list)
# print(*arr_copy)



```
